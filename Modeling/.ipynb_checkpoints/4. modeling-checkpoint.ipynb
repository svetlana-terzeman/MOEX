# Анализ по decile

```python
y_pred_prob_build = model.predict_proba(build[features])[:,1]
# 1. Собираем DataFrame
df_lift = pd.DataFrame({
    'y_true': build[target].values,
    'y_score': y_pred_prob_build
})

# 2. Сортируем по скору (по убыванию)
df_lift = df_lift.sort_values('y_score', ascending=False).reset_index(drop=True)

# 3. Добавляем номер дециля (1 = top 10%)
df_lift['decile'] = pd.qcut(
    df_lift.index,
    q=20,
    labels=False
) + 1

# 4. Базовый CTR по всему тесту
base_ctr = df_lift['y_true'].mean()

# 5. Агрегация по децилям
decile_stats = (
    df_lift
    .groupby('decile')
    .agg(
        mean_score  = ('y_score', 'mean'),
        min_score   = ('y_score', 'min'),
        max_score   = ('y_score', 'max'),
        bids = ('y_true', 'count'),
        clicks      = ('y_true', 'sum'),     
    )
    .reset_index()
    .sort_values('decile')
)

# 6. CTR и Lift
decile_stats['ctr'] = decile_stats['clicks'] / decile_stats['bids']
decile_stats['lift'] = decile_stats['ctr'] / base_ctr

# 7. Cumulative Gain
total_clicks = decile_stats['clicks'].sum()
decile_stats['cum_clicks'] = decile_stats['clicks'].cumsum()
decile_stats['gain'] = decile_stats['cum_clicks'] / total_clicks

# 8. Красивое форматирование
decile_stats['ctr'] = decile_stats['ctr'].round(4)
decile_stats['lift'] = decile_stats['lift'].round(3)
decile_stats['gain'] = decile_stats['gain'].round(3)
decile_stats['mean_score'] = decile_stats['mean_score'].round(2)
decile_stats['min_score'] = decile_stats['min_score'].round(2)
decile_stats['max_score'] = decile_stats['max_score'].round(2)
decile_stats_build = decile_stats.copy()
decile_stats_build
```

```python
bins = (
    decile_stats_build
    .sort_values('decile')
    [['decile', 'min_score', 'max_score', 'ctr', 'lift']]
    .reset_index(drop=True)
)

bins.loc[bins['max_score'].idxmax(), 'max_score'] = 1
bins.loc[bins['min_score'].idxmin(), 'min_score'] = 0
intervals = pd.IntervalIndex.from_arrays(
    bins['min_score'],
    bins['max_score'],
    closed='left'
)
intervals[:10]
```

```python
scoring_lift = bins[['min_score', 'max_score', 'lift']]
```

```python
y_pred_prob_test = model.predict_proba(test[features])[:, 1]

df_test = pd.DataFrame({
    'y_true': test[target].values,
    'y_score': y_pred_prob_test
})

df_test['decile'] = pd.cut(
    df_test['y_score'],
    bins=intervals
)

# вместо интервалов подставляем номер дециля
df_test['decile'] = df_test['decile'].map(dict(zip(intervals, bins['decile'])))

# 5. Агрегация по децилям
decile_stats = (
    df_test
    .groupby(['decile'])
    .agg(

        min_score   = ('y_score', 'min'),
        max_score   = ('y_score', 'max'),
        mean_score  = ('y_score', 'mean'),
        bids = ('y_true', 'count'),
        clicks = ('y_true', 'sum'), 
        
    )
    .reset_index()
    .sort_values('decile')
)
decile_stats      = decile_stats_build[['decile', 'lift']].merge(decile_stats)
decile_stats.columns = ['decile', 'lift',  'min_score', 'max_score', 'mean_score', 'bids', 'clicks']
# 6. CTR и Lift
decile_stats['ctr_test'] = decile_stats['clicks'] / decile_stats['bids']
decile_stats['lift_test'] = decile_stats['ctr_test'] / base_ctr

# 7. Cumulative Gain
total_clicks = decile_stats['clicks'].sum()
decile_stats['cum_clicks'] = decile_stats['clicks'].cumsum()
decile_stats['gain'] = decile_stats['cum_clicks'] / total_clicks

# 8. Красивое форматирование
decile_stats['ctr_test'] = decile_stats['ctr_test'].round(4)
decile_stats['lift_test'] = decile_stats['lift_test'].round(3)
decile_stats['gain'] = decile_stats['gain'].round(3)
decile_stats['mean_score'] = decile_stats['mean_score'].round(2)
decile_stats['min_score'] = decile_stats['min_score'].round(2)
decile_stats['max_score'] = decile_stats['max_score'].round(2)
decile_stats_test = decile_stats.copy()
decile_stats_test
```

```python
def result_calc(df, model, features, target, intervals, bins):
    '''Функция для подсчета метрик для анализа модели по интервалам
         param: df    : выборка, для которой считаеются метрики
         param: new_row: словарь с данными по выборке
         param: target : название модели, по которой идет анализ
         param: flag   : включаются ли предыдущие поколения в выборку 
         return: датафрейм с метриками
    '''
    #формирование интервалов
    df['predict_proba']   = model.predict_proba(df[features])[:, 1]
    
    #создание интервалов
    df['decile']   = pd.cut(df['predict_proba'], bins=intervals)
    df['decile']   = df['decile'].map(dict(zip(intervals, bins['decile'])))
    result         = df.groupby('decile', as_index=False).agg(  min_score   = ('predict_proba', 'min'),
                                                                max_score   = ('predict_proba', 'max'),
                                                                mean_score  = ('predict_proba', 'mean'),
                                                                bids = (target,'count')
                                                             )

    result['mean_score'] = result['mean_score'].round(2)
    result['min_score'] = result['min_score'].round(2)
    result['max_score'] = result['max_score'].round(2)
    #подсчет других показателей
    result['ratio_bids']   = (result['bids']/result['bids'].sum()).round(3)
    
    #разворачиваем датафрем
    result  = result.iloc[::-1]
    result['bids_cumsum'] = result['bids'].cumsum()
    result['ratio_bids_cumsum'] = (result['ratio_bids'].cumsum()).round(2)
    #разворачиваем датафрейм в исходный вид
    result  = result.iloc[::-1]
    
    return result
```

# make_proba

```python
def create_make_proba(build, model_base, features, target, cumsum = True):
    '''Функция для создания словаря с данными вероятности на build для каждого интервала по всем моделям

         param: normalize: default=False: нормализовать ли справочник от 0 до 1
         return: словарь первичный ключ - название модели, вторичный - интервал, значение - вероятность
    '''
    make_proba = {}

    #поиск оптимального трешхолда
    y_pred_proba          = model_base.predict_proba(build[features])[:, 1] 
    optimal_threshold, *_ = find_best_threshold(build[target], y_pred_proba)
    optimal_threshold     = np.round(optimal_threshold, 2)

    y_pred                = [1  if x >= optimal_threshold else 0 for x in y_pred_proba]
    build['predict_proba']= y_pred_proba 
    build['predict']      = y_pred 

    #создание интервалов
    bins                  = [i/100 for i in range(0, 101, 5)] 
    labels                = [f"{bins[i]} - {bins[i+1]}" for i in range(len(bins)-1)]
    build['intervals']    = pd.cut(build['predict_proba'], bins=bins, labels=labels, right=False)

    #подсчет вероятности оттока в интервале
    result         = build.groupby('intervals', as_index=False).agg({target: [ 'count', 'sum']})
    result.columns = ['Бин', 'Кол-во примеров', 'Кол-во таргета']
    if cumsum:
        result['Кол-во примеров']        = result['Кол-во примеров'].cumsum()
        result['Кол-во таргета']         = result['Кол-во таргета'].cumsum()
    result['Доля таргета в потоке'] = np.round(result['Кол-во таргета']/result['Кол-во примеров'],4)

    #формирование итогового словаря
    make_proba[target]  = result.set_index('Бин')['Доля таргета в потоке'].to_dict()
    return make_proba

```

# Подготовка данных для MLFLOW

```python
def scoring_dict_func(model, sample, target, name):

    # итоговый словарь
    result_dict = {}
    # находим вероятности
    scores = pd.Series(model.predict_proba(sample[model.feature_names_])[:,1])

    # Определяем границы диапазонов с шагом 0.05
    bins = np.arange(0, 1.05, 0.05)

    # Разбиваем данные на диапазоны и считаем количество элементов в каждом
    counts = pd.cut(scores, bins=bins, include_lowest=True).value_counts().sort_index()

    # Вычисляем доли
    proportions = round(counts / len(scores), 3)

    # Создаем словарь с долями
    score_dict = {
        f"{interval.left:.2f}-{interval.right:.2f}": proportion
        for interval, proportion in proportions.items()
    }

    # добавляем данные в словарь
    result_dict['score'] = score_dict
    result_dict['target'] =  round(sample[target].mean(), 3)

    return result_dict
```
