```python
def add_datetime_features(df, col='datetime'):
    """
    Добавляет фичи из datetime-колонки в датафрейм:
    - day_of_week (0=понедельник)
    - day_name (название дня)
    - hour (час)
    - time_of_day (morning/afternoon/evening/night)
    - is_weekend (флаг выходного)
    - month (номер месяца)
    - season (зима/весна/лето/осень)
    - is_month_start (флаг начала месяца)
    - is_month_end (флаг конца месяца)
    """
    df = df.copy()
    df[col] = pd.to_datetime(df[col])

    df['day_of_week'] = df[col].dt.dayofweek
    df['day_name'] = df[col].dt.day_name()
    df['hour'] = df[col].dt.hour

    def get_time_of_day(hour):
        if 5 <= hour < 12:
            return "morning"
        elif 12 <= hour < 17:
            return "afternoon"
        elif 17 <= hour < 22:
            return "evening"
        else:
            return "night"

    def get_season(month):
        if month in [12, 1, 2]:
            return "winter"
        elif month in [3, 4, 5]:
            return "spring"
        elif month in [6, 7, 8]:
            return "summer"
        else:
            return "autumn"

    df['time_of_day'] = df['hour'].apply(get_time_of_day)
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
    df['month'] = df[col].dt.month
    df['season'] = df['month'].apply(get_season)
    df['is_first_half'] = (df[col].dt.day <= 15).astype(int)

    return df
```

```python
df_feat = add_datetime_features(df_feat)
```

```python
info = ['ticker', 'name', 'datetime']
features = ['open', 'high', 'low', 'close', 'volume',
       'rsi', 'macd', 'macd_signal', 'macd_diff', 'bb_bbh', 'bb_bbl', 'bb_bbm',
       'bb_bbp', 'ma_5', 'ma_20', 'daily_return', 'volatility_5d',
       'close_lag_1', 'close_lag_2', 'close_lag_3', 'close_lag_5',
       'close_lag_10', 'day_of_week','is_weekend', 'month', 'is_first_half']
cat_features = ['ticker', 'day_name', 'season']
target =['target']
```
