## Анализ вещественных фичей

```python
feature_reasons = {}
low_var_features                         = remove_low_variance_features(df, num_features, threshold=0.1)
feature_reasons['low_variance_features'] = low_var_features
num_features                             = list(set(num_features) - set(low_var_features))
```

```python
# удаление константных фичей
const_features                    = const_feature(df, num_features,  0.9)
num_features                      = list(set(num_features) - set(const_features))
feature_reasons['const_features'] = const_features

# describe фичей, у которых доля константного значения N 
if len(const_features) > 0:
    display(df[const_features].describe(percentiles=[0.80,0.9,0.95,0.99]))
```

```python
# удаление фичей с пропусками
nan_features                    = nan_values(df, num_features, 0.2)
num_features                    = list(set(num_features) - set(nan_features))
feature_reasons['nan_features'] = nan_features

# кол-во пропусков по данным фичам
if nan_features:
    (df[nan_features].isna().sum() / df.shape[0]).sort_values(ascending=False)
```

## Анализ Категориальных/Бинарных фичей

```python
if len(cat_features) > 0:
    const_cat_features                    = const_feature(df, cat_features,  0.95)
    cat_features                          = list(set(cat_features) - set(const_cat_features))
    feature_reasons['const_cat_features'] = const_cat_features


# describe фичей, у которых доля константного значения N 
if const_cat_features:
    display(df[const_cat_features.keys()].describe(percentiles=[0.80,0.9,0.95,0.99]))
```

```python
if len(cat_features) > 0:
    # удаление фичей N% пропусков в данных
    nan_cat_features                    = nan_values(df, cat_features, 0.2)
    cat_features                        = list(set(cat_features) - set(nan_cat_features))
    feature_reasons['nan_cat_features'] = nan_cat_features
    
if len(nan_cat_features) > 0:
    # кол-во пропусков по данным фичам
    display((df[nan_cat_features].isna().sum() / df.shape[0]).sort_values(ascending=False))
```

```python
reasons = []
for reason, features in feature_reasons.items():
    for feature, explanation in features.items():
        reasons.append({
            "reason"     : reason,
            "feature"    : feature,
            "explanation": explanation
        })

reasons_df = pd.DataFrame(reasons)
```

# Удаляем редки записи в выборке 

```python
def filter_rare_values(df: pd.DataFrame, cols: list[str], target, threshold: float = 0.005):
    """
    Удаляет строки с редкими значениями по указанным столбцам.

    Args:
        df (pd.DataFrame): входной датафрейм
        cols (list[str]): список столбцов для фильтрации
        threshold (float): минимальная доля встречаемости (по умолчанию 0.005 = 0.5%)

    Returns:
        tuple:
            - pd.DataFrame: очищенный датафрейм
            - dict: словарь {col: {value: доля}} с редкими значениями
    """
    df_filtered = df.copy()
    rare_info = {}

    for col in cols:
        value_counts = df[col].value_counts(normalize=True)
        
        # выделяем редкие значения
        rare_values = value_counts[value_counts < threshold]
        
        if not rare_values.empty:
            # сохраняем инфу о редких значениях
            rare_info[col] = rare_values.to_dict()

            display(f"Кол-во 1 в пропущенный значениях по {col}: {df_filtered[df_filtered[col].isin(rare_values.index)][target].sum()}")
            display(f"Доля 1 в пропущенный значениях по {col}: {df_filtered[df_filtered[col].isin(rare_values.index)][target].mean()}")
            # удаляем строки с этими значениями
            df_filtered = df_filtered[~df_filtered[col].isin(rare_values.index)]

    return df_filtered, rare_info
```
